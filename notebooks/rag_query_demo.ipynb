```python
# notebooks/02_rag_query_demo.ipynb
# This notebook demonstrates retrieving similar strategies using the RAG query pipeline

# Cell 1: Imports
import sys
import os
sys.path.append(os.path.abspath('../src'))

from rag_query import StrategyRetriever

# Cell 2: Sample strategy summary to query with
query_text = """
This strategy combines an Alpha signal with a Hedge overlay. It spans 2009 to 2019 and yields a total return near 40%.
It aims to reduce market exposure while maintaining edge from a predictive alpha model.
"""

# Cell 3: Initialize retriever and query
retriever = StrategyRetriever(
    index_path='/Users/abhisheksuresh/Documents/Strategy_Simulator_Agent/LLM_Strategy_Agent/faiss_index.index',
    meta_path='/Users/abhisheksuresh/Documents/Strategy_Simulator_Agent/LLM_Strategy_Agent/strategy_metadata.pkl'
)


results = retriever.query(query_text, top_k=3)

# Cell 4: Display retrieved strategies
for i, res in enumerate(results):
    print(f"Top {i+1} Match")
    print("Similarity:", res['similarity'])
    for k, v in res['metadata'].items():
        print(f"{k}: {v}")
    print("---")
```
